```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Stock.csv') #importing stock dataset shape(1989)
print("Shape = {},{}".format(df.shape[0],df.shape[1]))
```

```python
df.head() #Peak
```

```python
df.tail() #Peak again
```

```python
df = df.iloc[:,1].values  #only opening values
```

```python
df.shape                  #1D numpy array
```

```python
x = np.arange(1989) #[0,1,2,3,4,5...,1989]
```

```python
plt.plot(x,df,color= 'r')
```

```python
beta = 0.92
a=np.arange(1) # {v=np.arange(1)}=={[0,++]}
for i in range(1,1989):
    choice = beta*a[i-1]+(1-beta)*df[i] #
    a = np.append(a,[choice])
beta = 0.8
b=np.arange(1) # {v=np.arange(1)}=={[0,++]}
for i in range(1,1989):
    choice = beta*b[i-1]+(1-beta)*df[i] #
    b = np.append(b,[choice])
beta = 0.5
c=np.arange(1) # {v=np.arange(1)}=={[0,++]}
for i in range(1,1989):
    choice = beta*c[i-1]+(1-beta)*df[i] #
    c = np.append(c,[choice])
```

```python
print(a)
print(b)
print(c)
```

```python
#less value of beta means averaging over less no of days == looks more precise
#more value of beta means averaging over more no of days == looks more distorted
plt.plot(x[:500],df[:500],color='r')
plt.plot(x[100:500],a[100:500],color='b') #blue ---->beta = 0.92
plt.plot(x[100:500],b[100:500],color='y') #yellow--->beta = 0.8
plt.plot(x[100:500],c[100:500],color='m') #magenta-->beta = 0.5
plt.show()
```

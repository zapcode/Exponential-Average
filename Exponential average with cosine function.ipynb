```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
a = []
for i in np.arange(0,20,0.2):
    a.append(math.cos(i))
a = np.array(a)
a.size
```

```python
plt.scatter(np.arange(0,100),a,color='r')
```

```python
def expon(beta,arr):
    ex = [0]
    size = arr.size
    for i in range(1,size):
        choice = beta*ex[i-1]+(1-beta)*arr[i]
        ex.append(choice)
    ex = np.array(ex)
    return ex
```

```python
zero_point_five = expon(0.5,a)
zero_point_six = expon(0.6,a)
zero_point_seven = expon(0.7,a)
zero_point_eight = expon(0.8,a)
zero_point_nine = expon(0.9,a)
zero_point_nine_one = expon(0.91,a)
zero_point_nine_two = expon(0.92,a)
zero_point_nine_three = expon(0.93,a)
zero_point_nine_four = expon(0.94,a)
```

```python
plt.scatter(np.arange(0,100),a,color='r')
plt.plot(np.arange(0,100),zero_point_five,color = 'b')
plt.plot(np.arange(0,100),zero_point_six,color = 'y')
plt.plot(np.arange(0,100),zero_point_seven,color = 'b')
plt.plot(np.arange(0,100),zero_point_eight,color = 'b')
plt.plot(np.arange(0,100),zero_point_nine,color = 'b')
plt.plot(np.arange(0,100),zero_point_nine_one,color = 'b')
plt.plot(np.arange(0,100),zero_point_nine_two,color = 'b')
plt.plot(np.arange(0,100),zero_point_nine_three,color = 'b')
plt.plot(np.arange(0,100),zero_point_nine_four,color = 'b')
```
